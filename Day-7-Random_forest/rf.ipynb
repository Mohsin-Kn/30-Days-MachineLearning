```python
data =pd.get_dummies(data=df,drop_first=True)
```

```python
data.card_yes.unique()
```

```python
data
```

```python
X=data.drop(['card_yes'],axis=1)
y=data["card_yes"]
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_test,X_Train,y_test,y_train=train_test_split(X,y,random_state=42,test_size=.3)
```

```python
X_test
```

```python
X_Train
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf =RandomForestClassifier(n_estimators=100,max_features='sqrt',n_jobs=-1)
```

```python
rf.fit(X_Train,y_train)
```

```python
pred=rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report, f1_score,accuracy_score
```

```python
print(classification_report(pred,y_test))
```

```python
print(accuracy_score(pred,y_test)) #checking accuracy 
```

```python
print(f1_score(pred,y_test))
```

## Hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [50, 100, 200],       # Number of trees in the forest
    'max_depth': [10, 20, None],          # Maximum depth of the tree
    'min_samples_split': [2, 5, 10],      # Minimum number of samples required to split a node
    'min_samples_leaf': [1, 2, 4],        # Minimum number of samples required at each leaf node
    'max_features': ['sqrt', 'log2'],     # Number of features to consider at each split
}

grid=GridSearchCV(estimator=rf,param_grid=param_grid,cv=5,verbose=2,n_jobs=-1)
```

```python
grid.fit(X_Train, y_train)
```

```python
print(grid.best_estimator_)
```

```python
print(grid.best_params_)
```

```python
print(grid.best_score_)
```

```python
print(grid.best_index_)
```

```python
best_rf_model= grid.best_estimator_
```

```python
prediction = best_rf_model.predict(X_test)
```

```python
print(accuracy_score(prediction,y_test))
```
