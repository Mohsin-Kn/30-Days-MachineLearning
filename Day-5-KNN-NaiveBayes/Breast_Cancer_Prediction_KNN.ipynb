```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score,KFold,train_test_split
from sklearn.metrics import accuracy_score
```

```python
X=df1.drop("diagnosis",axis=1)
y=df1["diagnosis"]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X.shape
```

```python
y.shape
```

# cross validation without scaling

```python
knn=KNeighborsClassifier(n_neighbors=5)
kf=KFold(n_splits=10,shuffle=True,random_state=42)
cv_scores = cross_val_score(knn, X, y, cv=kf, scoring='accuracy')
```

```python
cv_scores
```

```python
print(f'Cross-Validation Scores: {cv_scores}')
```

```python
print(f'Average Accuracy: {cv_scores.mean()}')
print(f'Standard Deviation: {cv_scores.std()}')
```

# cross validation with scaling

```python
model=KNeighborsClassifier(n_neighbors=5)
kfold=KFold(n_splits=10,shuffle=True,random_state=42)
cv = cross_val_score(model, X_scaled, y, cv=kfold, scoring='accuracy')
```

```python
cv
```

```python
print(f'Average Accuracy: {cv.mean()}')
print(f'Standard Deviation: {cv.std()}')
```

```python
plt.figure(figsize=(10,6))
plt.plot(np.arange(1, len(cv) + 1), cv, marker='o', linestyle='-', color='b', label='Accuracy per Fold')
plt.title('Cross-Validation Accuracy for Each Fold')
plt.xlabel('Fold Number')
plt.ylabel('Accuracy')
plt.xticks(np.arange(1, len(cv) + 1))  # Set x ticks as fold numbers
plt.ylim([0.0, 1.0])  # Accuracy range from 0 to 1
plt.grid(True)
plt.legend()
plt.show()
plt.savefig("accuracy.png")
```

```python
k_values = range(1, 21)  # Test k values from 1 to 20
cv_scores_mean = []
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    kf = KFold(n_splits=10, shuffle=True, random_state=42)
    cv = cross_val_score(knn, X_scaled, y, cv=kf, scoring='accuracy')
    cv_scores_mean.append(np.mean(cv))

# Plot the mean cross-validation accuracy for each k
plt.figure(figsize=(10, 6))
plt.plot(k_values, cv_scores_mean, marker='o', linestyle='-', color='b', label='Mean Accuracy per k')
plt.title('KNN Cross-Validation Accuracy for Different k Values')
plt.xlabel('k Value')
plt.ylabel('Mean Accuracy')
plt.xticks(k_values)
plt.grid(True)
plt.legend()
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
k=3
knn=KNeighborsClassifier(n_neighbors=k)
knn.fit(X_train,y_train)
prediction = knn.predict(X_test)
```

```python
acc=accuracy_score(prediction,y_test)
print("Model Accuracy: {:.2f}%".format(acc * 100))
```
