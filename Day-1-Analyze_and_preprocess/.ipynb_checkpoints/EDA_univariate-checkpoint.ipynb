```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Univariate Data Analysis:
 <p> Univariate analysis involves the examination of a single variable. It's the simplest form of data analysis, where only one characteristic or attribute of the data set is analyzed. The main purpose of univariate analysis is to describe the data and find patterns that exist within it.</p>
 <p>   
     
     Histogram:
        Shows the distribution of a single variable.
        Used to visualize the frequency or probability distribution of a    continuous variable.

    Box Plot (Box and Whisker Plot):
        Summarizes the distribution of a single continuous variable.
        Displays the median, quartiles, and outliers.

    Bar Plot (Bar Chart):
        Displays the distribution of a categorical variable.
        Used to compare the frequencies of categories.
     
 </p>

 ## Mathematical Example:

```python
ages = np.array([25, 30, 35, 40, 45, 50, 55, 60, 65, 70])
print(np.mean(ages))
print(np.median(ages))
print(np.argmax(np.bincount(ages)))
print(np.max(ages)-np.min(ages))
print(np.std(ages))
```

### Univariate Graph (Histogram):

```python
# plt.figure(figsize=(8,6))
ages = np.array([25, 30, 35, 40, 45, 50, 55, 60, 65, 70])
plt.hist(ages,bins=10,color="green",)
```

## Univariate with Titanic 
### Categorical Data using Single column

```python
df =pd.read_csv("titanic.csv")
```

```python
df.head(2)
```

```python
#using countplot

sns.countplot(x=df["Embarked"])
```

```python
#using bar plot 
df["Survived"].value_counts().plot(kind="bar")
```

```python
# using pie chart

df["Sex"].value_counts().plot(kind="pie",autopct="%.2f")
```

```python
x=[33,57,100]
y=["a","b","c"]
c=["red","yellow","green"]
plt.pie(x=x,labels=y,autopct="%.2f",colors=c)
plt.show()
```
